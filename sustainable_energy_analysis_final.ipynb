# Energy Consumption Prediction with Machine Learning and Deep Learning Models



This project implements three models‚Äî**CNN (Convolutional Neural Network)**, **RNN (Recurrent Neural Network)**, and **Random Forest Regressor**‚Äîto predict **Energy Consumption (MWh)** based on the features in the dataset.

---

### Steps and Methods

1. **Data Preprocessing**:
   - The dataset is loaded, and the target variable (`Energy_Consumption_MWh`) is separated.
   - Features are normalized using `StandardScaler` to standardize the input for machine learning and deep learning models.
   - Data is split into training (80%) and testing (20%) sets.

2. **Deep Learning Models**:
   - **CNN**:
     - Uses two 1D convolutional layers followed by max pooling, flattening, and dense layers.
     - A dropout layer is added for regularization.
   - **RNN**:
     - Built with two LSTM layers to capture sequential dependencies, followed by dense and dropout layers.
   - Both models are compiled with the **Adam optimizer** and trained for **20 epochs**.

3. **Random Forest Regressor**:
   - A traditional ensemble method with 100 decision trees is used as a baseline.
   - Predictions are evaluated using **Mean Squared Error (MSE)**, **Mean Absolute Error (MAE)**, and **R¬≤ Score**.

4. **Evaluation and Visualization**:
   - Training and validation loss for CNN and RNN models are plotted to observe learning behavior over epochs.
   - Model evaluation metrics (MSE and MAE) are reported for all three models.

---

### Results and Observations

1. **Random Forest Regressor**:
   - **MSE**: Indicates the average squared difference between actual and predicted values.
   - **MAE**: Represents the average absolute error.
   - **R¬≤ Score**: Shows how well the model explains the variance in the data.

2. **CNN and RNN**:
   - Loss plots are generated for training and validation sets, showcasing model convergence and overfitting behavior if present.
   - Final evaluation metrics (MSE and MAE) are computed on the test set.

3. **Comparison**:
   - All models are compared based on their performance metrics, providing insights into the most effective approach for predicting energy consumption.

---

### Key Insights
- **Random Forest** offers a robust baseline with quick implementation.
- **Deep Learning Models** (CNN and RNN) are more complex and suitable for datasets with hidden patterns or sequential dependencies.
- Visualizing loss trends helps in diagnosing overfitting or underfitting in deep learning models.

This analysis provides a comprehensive framework for predictive modeling of energy consumption.

```python
# Required Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, MaxPooling1D, Dropout, LSTM

# Load Dataset
file_path = 'updated_energy_dataset.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Features and Target
X = data.drop(columns=["Energy_Consumption_MWh"])
y = data["Energy_Consumption_MWh"]

# Normalize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Prepare Data for Deep Learning Models (CNN, RNN)
X_train_reshaped = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_reshaped = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# 1. CNN Model
cnn_model = Sequential([
    Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Conv1D(32, kernel_size=3, activation='relu'),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1)
])
cnn_model.compile(optimizer='adam', loss='mse', metrics=['mae'])
cnn_history = cnn_model.fit(X_train_reshaped, y_train, validation_data=(X_test_reshaped, y_test),
                            epochs=20, batch_size=32, verbose=1)

# 2. RNN Model
rnn_model = Sequential([
    LSTM(64, activation='relu', return_sequences=True, input_shape=(X_train.shape[1], 1)),
    LSTM(32, activation='relu'),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1)
])
rnn_model.compile(optimizer='adam', loss='mse', metrics=['mae'])
rnn_history = rnn_model.fit(X_train_reshaped, y_train, validation_data=(X_test_reshaped, y_test),
                            epochs=20, batch_size=32, verbose=1)

# 3. Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)

# Random Forest Evaluation
rf_mse = mean_squared_error(y_test, rf_predictions)
rf_mae = mean_absolute_error(y_test, rf_predictions)
rf_r2 = r2_score(y_test, rf_predictions)
print(f"Random Forest - MSE: {rf_mse}, MAE: {rf_mae}, R2: {rf_r2}")

# Plot Training and Validation Loss for CNN and RNN
def plot_history(history, title):
    plt.figure(figsize=(8, 5))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f"{title} Loss")
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

plot_history(cnn_history, "CNN")
plot_history(rnn_history, "RNN")

# Evaluate CNN and RNN Models
cnn_loss, cnn_mae = cnn_model.evaluate(X_test_reshaped, y_test, verbose=0)
rnn_loss, rnn_mae = rnn_model.evaluate(X_test_reshaped, y_test, verbose=0)

print(f"CNN - MSE: {cnn_loss}, MAE: {cnn_mae}")
print(f"RNN - MSE: {rnn_loss}, MAE: {rnn_mae}")
```
